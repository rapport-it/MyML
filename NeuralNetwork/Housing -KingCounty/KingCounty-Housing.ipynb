```python
#start dropiing the colum that is not usefull
df = df.drop('id', axis = 1)
```

```python
 
df['date']
```

```python
# start working on date time
# date is string object 
# start convert to date object so ve can extract month and date 
df['date'] = pd.to_datetime(df['date'])
```

```python
# we got datatime object 
df['date']
```

```python
#extract year 
df['year'] = df['date'].apply(lambda date: date.year  )
# Extract month 
df['month'] = df['date'].apply(lambda date: date.month  )
```

```python
# drop date 
df = df.drop('date', axis = 1)
# drop zip
df = df.drop('zipcode', axis = 1)
```

```python
df.groupby('month').mean()['price'].plot()
```

```python
df.groupby('year').mean()['price'].plot()
# due to inflasion   
```

```python
df['sqft_basement'].value_counts()
```

```python
X= df.drop('price', axis = 1).values
y = df['price'].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
X_train = scaler.fit_transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
# look at the feature 
X_train.shape
```

```python
model = Sequential()
model.add(Dense(19, activation = 'relu'))
model.add(Dense(19, activation = 'relu'))
model.add(Dense(19, activation = 'relu'))
model.add(Dense(19, activation = 'relu'))
model.add(Dense(1))
model.compile( optimizer ='adam', loss = 'mse')
```

```python
model.fit(x= X_train, y= y_train,validation_data= (X_test, y_test), batch_size= 128, epochs =400 )
```

```python
losses = pd.DataFrame( model.history.history)
```

```python
losses.plot()
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, explained_variance_score
```

```python
prediction = model.predict(X_test)
```

```python
mean_squared_error(y_test, prediction)
```

```python
# root SQure Error 
np.sqrt(mean_squared_error(y_test, prediction))
```

```python
mean_absolute_error(y_test, prediction)
```

```python
# see mean mean absolute error   
df['price'].describe()
```

```python
explained_variance_score(y_test, prediction)
```

```python
plt.figure(figsize= (12,6))
plt.scatter(y_test, prediction)
plt.plot(y_test, y_test,'r')
```

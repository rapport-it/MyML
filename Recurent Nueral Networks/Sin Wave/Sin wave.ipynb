```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,50,501)
```

```python
x
```

```python
y =np.sin(x)
```

```python
y
```

```python
plt.plot(x,y)
```

```python
df = pd.DataFrame(data=y, index = x, columns=['Sin'])
```

```python
df
```

```python
test_percentage = 0.1
```

```python
len(df)* test_percentage
```

```python
test_point = np.round(len(df)* test_percentage)
```

```python
test_point
```

```python
test_index = int(len(df) - test_point)
```

```python
test_index
```

```python
train = df.iloc[:test_index]
```

```python
test = df.iloc[test_index:]
```

```python
train
```

```python
test
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scalar = MinMaxScaler()
```

```python
scaled_train = scalar.fit_transform(train)
```

```python
scaled_test = scalar.transform(test)
```

```python
from tensorflow.keras.preprocessing.sequence import TimeseriesGenerator
```

```python
length = 25 # what is the given
batch_size = 2 # how many next
generator = TimeseriesGenerator(scaled_train, scaled_train, length=length,batch_size=batch_size)
```

```python
x,y = generator[0]
```

```python
x
```

```python
# continution call 
y
```

```python
length = 50 # what is the given
batch_size = 1 # how many next
generator = TimeseriesGenerator(scaled_train, scaled_train, length=length,batch_size=batch_size)
```

```python
from tensorflow.keras.models import Sequential
```

```python
from tensorflow.keras.layers import Dense, SimpleRNN, LSTM
```

```python
numFeatures = 1
```

```python
model = Sequential()
model.add(SimpleRNN(50, input_shape = (length, numFeatures) ))
model.add(Dense(1))
model.compile(optimizer ='adam', loss= 'mse')
```

```python
model.summary()
```

```python
model.fit_generator(generator, epochs=5)
```

```python
loses = pd.DataFrame(model.history.history)
```

```python
loses.plot()
```

```python
test_predication =[]

first_eval_batch = scaled_train[-length:]
current_batch = first_eval_batch.reshape(1,length, numFeatures)
for i in range(len(test)):
    currect_predication = model.predict(current_batch)[0]
    test_predication.append(currect_predication)
    current_batch = np.append(current_batch[:,1:, :], [[currect_predication]],axis=1)
```

```python
test_predication
```

```python
true_predication = scalar.inverse_transform(test_predication)
```

```python
test['Predication'] = true_predication
```

```python
test
```

```python
test.plot()
```

```python
from tensorflow.keras.callbacks import EarlyStopping
```

```python
early_stop = EarlyStopping(monitor= 'val_loss', patience= 2)
```

```python

genrator = TimeseriesGenerator(scaled_train,scaled_train, length=50,batch_size=1 )
validation_generator = TimeseriesGenerator(scaled_test, scaled_test, length=49, batch_size=1 )
```

```python
model = Sequential()
model.add(LSTM(50, input_shape = (49, numFeatures) ))
model.add(Dense(1))
model.compile(optimizer ='adam', loss= 'mse')
```

```python
model.fit_generator(generator, epochs=20, validation_data= validation_generator, callbacks=[early_stop])
```

# Basic Methods

```python
def priceToBook(summary_result):
    if not("defaultKeyStatistics" in summary_result.keys()):
        return 0
    if ("priceToBook" in summary_result["defaultKeyStatistics"].keys()) and ("raw" in summary_result["defaultKeyStatistics"]["priceToBook"].keys()) and summary_result["defaultKeyStatistics"]["priceToBook"]["raw"] is not None:
        return summary_result["defaultKeyStatistics"]["priceToBook"]["raw"]
    return 0.0
```

```python
def peValue(summary_result):
    trailingPE = {}
    if not("summaryDetail" in summary_result.keys()):
        return 0.0 
    if ("trailingPE" in summary_result["summaryDetail"].keys()) and ("raw" in summary_result["summaryDetail"]["trailingPE"].keys())and summary_result["summaryDetail"]["trailingPE"]["raw"] is not None:
        return summary_result["summaryDetail"]["trailingPE"]["raw"]
    return 0.0
```
